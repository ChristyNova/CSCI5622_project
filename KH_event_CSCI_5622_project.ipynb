```python
# Define the z-score function
def zscore(x, window):
    r = x.rolling(window=window)
    m = r.mean().shift(1)
    s = r.std(ddof=0).shift(1)
    z = (x-m)/s
    return z
    
# Calculate zscore with a rolling window and get abnormal values
window_size = 5 
df_zscore_n = zscore(MMS2_nnn.n, window_size)

times_n = pd.date_range('2020-06-17/20:00:00','2020-06-18/01:30:00',periods=len(df_zscore_n))


df_zscore_n1 = pd.DataFrame({'n':df_zscore_n,'times':times_n})
# Apply a threshold for the z-score set it to every ___ minutes
threshold = 15
abnormal_values_dates_n = df_zscore_n[abs(df_zscore_n1.n) > threshold ].index


# MMS2_nn.plot(figsize =(14,6))
df_zscore_Tperp = zscore(MMS2_Tperp2.Tperp, 4)

times_Tperp = pd.date_range('2020-06-17/20:00:00','2020-06-18/01:30:00',periods=len(df_zscore_Tperp))


df_zscore_Tperp1 = pd.DataFrame({'Tperp':df_zscore_Tperp,'times':times_Tperp})
# Apply a threshold for the z-score set it to every ___ minutes

threshold_Tperp = 20
abnormal_values_dates_Tperp = df_zscore_Tperp[abs(df_zscore_Tperp1.Tperp) > threshold_Tperp ].index

plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(20,6))
plt.plot(times_n, df_zscore_n)
plt.title('Outliers in Density dataset')
plt.xlabel('Time Stamp')
plt.ylabel('Z-score')
plt.axhline(y = threshold,color='purple',linestyle='-', label='threshold')
plt.axhline(y = -threshold,color='purple',linestyle='-')
plt.grid()
plt.legend()
plt.show




plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(20,6))
plt.plot(times_Tperp, df_zscore_Tperp)
plt.title('Outliers in $T_{⊥}$ dataset')
plt.xlabel('Time Stamp')
plt.ylabel('Z-score')
plt.axhline(y = threshold_Tperp,color='purple',linestyle='-', label='threshold')
plt.axhline(y = -threshold_Tperp,color='purple',linestyle='-')
plt.grid()
plt.legend()
plt.show()
```
